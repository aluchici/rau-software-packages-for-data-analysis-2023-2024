```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Spotify_Dataset_V3.csv", sep=";")
df.head()
```

```python
len(df)
```

```python
columns = df.columns
for column in columns:
  if type(df[column]) == int or type(df[column]) == float:
    missing_entries = np.isnan(df[column])
    missing_values = df[column][missing_entries]
    print(missing_values)
```

```python
df.dropna(inplace=True)
```

```python
avg = np.mean(df["Speechiness"])
low = np.quantile(df["Speechiness"], 0.01)
high = np.quantile(df["Speechiness"], 0.99)
iqr = high - low
min_value = avg - 1.5 * iqr
max_value = avg + 1.5 * iqr
min_value, max_value
```
