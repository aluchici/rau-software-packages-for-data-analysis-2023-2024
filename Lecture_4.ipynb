```python
import matplotlib.pyplot as plt

dt = pd.to_datetime(df["date"])
df["timestamp"] = dt.astype('int64') // 10**9

x = df[(df["country"] == "Finland") & (df["product"] == "Kaggle Hat") & (df["store"] == "KaggleMart")]["timestamp"][:30]
y = df[(df["country"] == "Finland") & (df["product"] == "Kaggle Hat") & (df["store"] == "KaggleMart")]["num_sold"][:30]
```

```python
_ = plt.figure(figsize=(20, 10))
plt.bar(x=x, height=y)
plt.show()
```

```python
_ = plt.figure(figsize=(20, 10))
plt.plot(x, y)
plt.scatter(x, y, s=100)
plt.show()
```

```python
x1 = x.values[0]
x2 = x.values[-1]

y1 = y.values[0]
y2 = y.values[-1]

a = (y1 - y2) / (x1 - x2)
b = y1 - a*x1

y_trend = []
for xi in x:
  yi = a * xi + b
  y_trend.append(yi)

```

```python
_ = plt.figure(figsize=(20, 10))
plt.plot(x, y)
plt.scatter(x, y, s=100)
plt.plot(x, y_trend)
plt.show()
```
